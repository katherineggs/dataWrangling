```python
import numpy as np
import pandas as pd
```

```python
# Leemos el csv
heroes = pd.read_csv("heroes_information.csv")
```

```python
# Comprobamos
heroes
```

```python
from pandasql import *

sqlDf = lambda q: sqldf(q,globals())
```

```python
# Primer nombre 
query = """ 
SELECT name FROM heroes 
LIMIT 1
"""

sqlDf(query)
```

```python
# Casas Publicadoras
query = """
SELECT Publisher, COUNT(*) FROM heroes
GROUP BY (Publisher)
"""

sqlDf(query)
```

```python
# CUANTOS Superheores de 2m
query = """
SELECT COUNT(*) FROM heroes
WHERE Height > 200
"""

sqlDf(query)
```

```python
# CUANTOS Superheores de 2m y humanos
query = """
SELECT COUNT(*) FROM heroes
WHERE (Height > 200) and (Race = "Human")
"""

sqlDf(query)
```

```python
# CUANTOS Peso mÃ¡s de 100 lbs y menos de 200 lbs
query = """
SELECT COUNT(*) FROM heroes
WHERE (Weight > 100) and (Weight < 200)
"""

sqlDf(query)
```

```python
# CUANTOS Ojos de color azul o rojo
query = """
SELECT "Eye color", COUNT(*) FROM heroes
WHERE "Eye color" IN ("red", "blue")
GROUP BY "Eye color"
"""

sqlDf(query)
```

```python
# CUANTOS Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro
query = """
SELECT name, Race, "Hair color" FROM heroes
WHERE Race IN ("Human", "Mutant") 
AND "Hair color" IN ("Green")

OR Race IN ("Vampire") 
AND "Hair color" IN ("Black")
"""

sqlDf(query)
```

```python
# Primer superhÃ©roe si ordeno la tabla por raza en orden descendente
query = """
SELECT name FROM heroes
ORDER BY Race DESC
limit 1
"""

sqlDf(query)
```

```python
# SuperhÃ©roes son de gÃ©nero masculino y cuÃ¡ntas de gÃ©nero femenino
query = """
SELECT Gender, COUNT(*) FROM heroes
GROUP BY Gender
"""

sqlDf(query)
```

```python
# Casas publicadoras tienen mÃ¡s de 15 superhÃ©roes
query = """
SELECT Publisher, COUNT(*) FROM heroes
GROUP BY Publisher 
HAVING COUNT(*) > 15
"""

sqlDf(query)
```
